```python
import pandas as pd
import numpy as np
from sklearn.metrics import classification_report
```

```python
import spacy
```

```python
nlp=spacy.load('en_core_web_sm')
```

```python
nlp.pipe_names
```

```python
text=nlp("Nvidia is going to launch a new GPU worth $10 billion")
for ent in text.ents:
  print(ent.text,"|",ent.label_,"|",spacy.explain(ent.label_))
```

```python
from spacy import displacy
displacy.render(text,style='ent')
```

```python
nlp.pipe_labels
```

```python
doc=nlp("Mark Zuckerberg founded Meta in 2010")
displacy.render(doc,style='ent')
```

```python
my_text="Navaraja is going to get a job in Riot Games as a Data Engineer"
doc=nlp(my_text)
displacy.render(doc,style='ent')
```

```python
from spacy.tokens import Span
```

```python
s1=Span(doc,0,1,label='PERSON')
s2=Span(doc,8,10,label='ORG')
s3=Span(doc,12,14,label='JOB')
doc.set_ents([s1,s2,s3],default='unmodified')
```

```python
for ent in doc.ents:
  print(ent.text,"|",ent.label_)
```
