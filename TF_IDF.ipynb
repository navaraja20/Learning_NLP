```python
import pandas as pd
import numpy as np
from sklearn.metrics import classification_report
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
```

```python
corpus=[
    "Ironman eating crossiant, Hulk is eating crossiant, Loki ate crossiant already",
    "Nvidia is announcing new GPU tomorrow",
    "Dodge is announcing new car model tomorrow",
    "Iqoo is announcing new phone model tomorrow",
    "Tesla is announcing new car model tomorrow",
    "I am eating biryani and you are eating grapes"
]
```

```python
v=TfidfVectorizer()
v.fit(corpus)
transform_output=v.transform(corpus)
#print(transform_output.toarray())
```

```python
print(v.vocabulary_)
```

```python
all_features_name=v.get_feature_names_out()
for word in all_features_name:
    index=v.vocabulary_.get(word)
    idf_score=v.idf_[index]
    print(f"{word}:{idf_score}")
```

```python
df=pd.read_csv('/content/Ecommerce_data.csv')
```

```python
df['label_num']=df['label'].map({
    'Household':0,
    'Electronics':1,
    'Clothing & Accessories':2,
    'Books':3
})
```

```python
df.head()
```

```python
X_train,X_test,y_train,y_test=train_test_split(df.Text,df.label_num,test_size=0.2)
```

```python
len(X_train)
```

```python
tf=TfidfVectorizer()
X_train_tf=tf.fit_transform(X_train)
X_test_tf=tf.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier()
clf.fit(X_train_tf,y_train)
y_pred=clf.predict(X_test_tf)
print(classification_report(y_test,y_pred))
```
